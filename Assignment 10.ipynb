```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec

jobs = [
    "python developer machine learning data science",
    "web developer html css javascript react",
    "data analyst sql excel visualization",
    "ai engineer deep learning neural networks",
    "backend developer nodejs api database"
]

resume = "python machine learning deep learning projects"
\
def tokenize(text):
    return text.lower().split()

corpus = [tokenize(j) for j in jobs]
corpus.append(tokenize(resume))

model = Word2Vec(corpus, vector_size=50, window=3, min_count=1)

def get_vector(text):
    words = tokenize(text)
    vecs = []
    for w in words:
        if w in model.wv:
            vecs.append(model.wv[w])
    if len(vecs) == 0:
        return np.zeros(50)
    return np.mean(vecs, axis=0)

resume_vec = get_vector(resume)

scores = []
for job in jobs:
    job_vec = get_vector(job)
    score = cosine_similarity([resume_vec], [job_vec])[0][0]
    scores.append((job, score))

scores = sorted(scores, key=lambda x: x[1], reverse=True)

for job, score in scores:
    print(f"{job} --> {score:.2f}")
```
