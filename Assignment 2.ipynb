```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

text = """Natural Language Processing helps machines understand human language.
Students are learning advanced NLP techniques in their college."""

words = word_tokenize(text)

pos_tags = pos_tag(words)

print("Words:")
print(words)

print("\nPOS Tags:")
print(pos_tags)
```
