```python
import spacy
import numpy as np
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    doc = nlp(text.lower())
    tokens = [
        token.lemma_ for token in doc
        if not token.is_stop and not token.is_punct
    ]
    return " ".join(tokens)
```

```python
greeting_keywords = {"hi", "hello", "hey", "good morning", "good evening"}
goodbye_keywords = {"bye", "goodbye", "see you"}
thanks_keywords = {"thanks", "thank you", "appreciate"}
help_keywords = {"help", "support", "assist"}

def detect_simple_intent(user_input):
    text = user_input.lower()

    for word in greeting_keywords:
        if word in text:
            return "greeting"

    for word in goodbye_keywords:
        if word in text:
            return "goodbye"

    for word in thanks_keywords:
        if word in text:
            return "thanks"

    for word in help_keywords:
        if word in text:
            return "help"

    return None
```

```python
knowledge_base = {
    "what is machine learning": "Machine learning is a field of AI that allows systems to learn from data.",
    "what is nlp": "Natural Language Processing enables computers to understand human language.",
    "who are you": "I am a simulated intelligent chatbot built using NLP techniques.",
    "what is artificial intelligence": "Artificial Intelligence is the simulation of human intelligence in machines.",
    "what is python": "Python is a high-level programming language widely used in AI and data science."
}

kb_questions = list(knowledge_base.keys())
kb_processed = [preprocess(q) for q in kb_questions]

kb_vectorizer = TfidfVectorizer()
kb_vectors = kb_vectorizer.fit_transform(kb_processed)
```

```python
def retrieve_answer(user_input):
    processed = preprocess(user_input)
    user_vector = kb_vectorizer.transform([processed])

    similarities = cosine_similarity(user_vector, kb_vectors)
    max_index = np.argmax(similarities)
    max_score = similarities[0][max_index]

    if max_score > 0.4:
        return knowledge_base[kb_questions[max_index]]
    else:
        return None
```

```python
generic_responses = [
    "That's interesting. Tell me more.",
    "Why do you think that?",
    "That sounds important.",
    "Can you explain that further?",
    "I see. Go on...",
    "Hmm, that's something to think about."
]

emotional_words = {
    "happy": "I'm glad to hear that!",
    "sad": "I'm sorry to hear that. Want to talk about it?",
    "angry": "That sounds frustrating.",
    "excited": "That’s awesome!",
    "tired": "Maybe you should take a short break."
}

def simulated_response(user_input):
    text = user_input.lower()

    for word in emotional_words:
        if word in text:
            return emotional_words[word]

    if "?" in user_input:
        return "That's a thoughtful question. What do you think about it?"

    if len(user_input.split()) > 8:
        return "That sounds quite detailed. I appreciate you sharing that."

    return random.choice(generic_responses)
```

```python
def chatbot_response(user_input):

    simple_intent = detect_simple_intent(user_input)

    if simple_intent == "greeting":
        return "Hello! How can I help you?"

    elif simple_intent == "goodbye":
        return "Goodbye! Have a great day!"

    elif simple_intent == "thanks":
        return "You're welcome!"

    elif simple_intent == "help":
        return "Sure! Tell me what you need help with."

    answer = retrieve_answer(user_input)
    if answer:
        return answer

    return simulated_response(user_input)
```

```python
print("Simulated AI Chatbot is ready! Type 'exit' to stop.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Goodbye!")
        break

    response = chatbot_response(user_input)
    print("Bot:", response)
```
