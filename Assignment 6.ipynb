```python
# ==============================
# TEXT CLASSIFICATION (50 DATASET)
# ==============================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# ------------------------------
# LARGE DATASET (50 sentences)
# ------------------------------
data = {
    "text": [
        # Positive (25)
        "I love this product",
        "This is amazing",
        "Very happy with the service",
        "Best experience ever",
        "Absolutely fantastic",
        "I really enjoyed this",
        "Superb quality",
        "Very satisfied",
        "Highly recommended",
        "Worth every penny",
        "Excellent performance",
        "Great customer support",
        "I am delighted",
        "Top quality product",
        "Loved it so much",
        "Perfect choice",
        "Very impressive",
        "Five stars experience",
        "Awesome service",
        "Really good",
        "Extremely happy",
        "Fantastic purchase",
        "Great job",
        "Amazing quality",
        "Totally worth it",

        # Negative (25)
        "I hate this",
        "Worst product",
        "Very bad experience",
        "Not satisfied at all",
        "Terrible service",
        "Completely disappointed",
        "Waste of money",
        "Poor quality",
        "Not worth it",
        "Very frustrating",
        "Bad customer support",
        "Awful experience",
        "I regret buying this",
        "Extremely bad",
        "Not recommended",
        "Very disappointing",
        "Horrible product",
        "Worst experience ever",
        "Really bad",
        "Totally useless",
        "Cheap quality",
        "Not good",
        "Very unhappy",
        "Disgusting service",
        "Pathetic performance"
    ],
    "label": [
        "positive"] * 25 + ["negative"] * 25
}

df = pd.DataFrame(data)

# ------------------------------
# SPLIT (STRATIFIED)
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    df["text"],
    df["label"],
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)

# ------------------------------
# TF-IDF
# ------------------------------
vectorizer = TfidfVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# ------------------------------
# MODEL
# ------------------------------
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# ------------------------------
# PREDICTION
# ------------------------------
y_pred = model.predict(X_test_tfidf)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

def predict(text):
    return model.predict(vectorizer.transform([text]))[0]

print("\nCustom Predictions:")
print("I love this →", predict("I love this"))
print("This is worst →", predict("This is worst"))
print("Amazing service →", predict("Amazing service"))
print("Very bad product →", predict("Very bad product"))
```
